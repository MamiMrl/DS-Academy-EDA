```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("nyc_census_tracts.csv")
```

```python
# Use df.info() to display information about the dataset, including column names, data types, and non-null values
df.info()
```

```python
# Use df.duplicated().sum() to count the number of duplicate rows in the dataset and print the result
print(df.duplicated().sum())
```

```python
# Use print(df.columns) to display a list of column names in the dataset
print(df.columns)
```

```python
# Listing the columns which are necessary
columns_to_keep = ['CensusTract', 'County', 'Borough', 'TotalPop', 'Men', 'Women', 'Hispanic', 'White', 'Native','Asian','Citizen','Income','IncomePerCap','Professional','SelfEmployed','Poverty']

# Dropping the columns which are unnecessary
df2 = df.drop(columns=[col for col in df.columns if col not in columns_to_keep])
```

```python
# Use df2.loc to create two new dataframes containing rows where 'Income' is null and columns for 'Women' and 'Income' or 'Men' and 'Income'
women_and_null_income = df2.loc[df2['Income'].isnull(), ['Women', 'Income']]
men_and_null_income = df2.loc[df2['Income'].isnull(), ['Men', 'Income']]

# Use print() to display the two new dataframes
print(women_and_null_income)
print(men_and_null_income)
```

```python
# Creating a box plot of the income column
fig, ax = plt.subplots()
ax.boxplot(df2['Income'])

# Setting the axis labels and title
ax.set_xlabel('Income')
ax.set_ylabel('Value')
ax.set_title('Box plot of Income column')

plt.show() 
```

```python
# Use sns.boxplot() to create a box plot of 'Income' from df2
ax = sns.boxplot(y = 'Income', data = df2)
```

```python
# Use df2['Income'].fillna() to fill missing values in the 'Income' column with a specific value
df2['Income'].fillna(59101.320800, inplace=True)
```

```python
print(df2.isnull().sum())
```

```python
# Use df2.groupby() and .agg() to group data by 'County' and 'Borough', and aggregate other columns by summing them up
grouped_data = df2.groupby(['County', 'Borough']).agg({'TotalPop': 'sum', 'Men': 'sum', 'Women': 'sum', 'Income': 'sum'}).reset_index()
```

```python
grouped_data
```

```python
# Use df2.groupby() to group data by 'Borough', calculate the mean income for each borough, and plot it using a bar chart
grouped_data = df2.groupby('Borough')['Income'].mean()

# Use sns.color_palette() to generate a list of colors, with one color for each borough
colors = sns.color_palette('husl', len(grouped_data))

# Use plt.bar() to plot the mean income for each borough, using the index of the 'grouped_data' dataframe as x-axis values
plt.bar(grouped_data.index, grouped_data.values, color=colors)

# Set the plot title, x-axis label, and y-axis label
plt.title('Mean Income by Borough')
plt.xlabel('Borough')
plt.ylabel('Income')

# Save the plot as a PNG image with high resolution, and show the plot
plt.savefig("NYC-Census-meanIncomeBorough.png", dpi = 600)
plt.show()
```

```python
grouped_data = df2.groupby('Borough')[['TotalPop', 'Men', 'Women']].sum()
sns.set_theme()
sns.set_style("whitegrid")
sns.set_context("paper")

plt.figure(figsize=(12,7))
grouped_data.plot(kind = 'bar')

plt.title('Population by Borough')
plt.xlabel('Borough')
plt.ylabel('Population')

ymax = 2700000

#plt.ylim(ymin, ymax)

plt.ylim([0, int(ymax)])
plt.show()
```
