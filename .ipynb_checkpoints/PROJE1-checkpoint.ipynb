### <font color="blue">Tekrar eden verileri verisetimizden temizliyoruz.</font>

### <font color="blue">C/A, UNIT, SCP, STATION sütunları her turnike için spesifik olmalı. Bundan dolayı turnike bazlı giriş çkışaları kontrol edebilmek için bu dört değerin ortak olduğu "turnstile" sütunu oluşturduk.

```python
df["turnstile"] = list([i[0]+"*"+i[1]+"*"+i[2]+"*"+i[3] for i in zip(df["C/A"],df.UNIT,df.SCP,df.STATION)])
```

### <font color="blue"> Entries ve exits değerlerimiz cumulative sum olduğu için her saat aralığında geçen net insan sayısını sayısını belirlemek için net_entries ve net_exits sütunlarını oluşturduk.  

```python
df["net_entries"]=df.ENTRIES.shift(-1) - df.ENTRIES
df["net_exits"] = df.EXITS.shift(-1) - df.EXITS
```

```python
df
```

```python
df.describe()
```

<font color="blue"> net entries ve net exits sütunlarımıza baktığımızda, minimum değer negatif olmamalıydı. maximum değerlerimiz de bu derece büyük olmamalı. O zaman bu aykırı değerlerimiz nelermiş ve kaç taneymiş kontrol edelim.

### <font color="blue">Her saniyede bir kişi geçtiğini varsaysak 4 saatlik aralıklarda maximum 4x60x60=14400 kişi geçebilir.

```python
df[(df.net_entries<0) | (df.net_entries>14400) | (df.net_exits<0) | (df.net_exits>14400)]
```

### <font color="blue">75286 satır aykırı değerlere sahip. Yani toplam 2112631 satır verimizin yaklaşık %3.56'sı aykırı değerlere sahip. Ve bu miktarın silineblir olduğunu düşünerek bu satırları verimizden siliyoruz. 

```python
df.drop(df[(df.net_entries<0) | (df.net_entries>14400) | (df.net_exits<0) | (df.net_exits>14400)].index,inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
df.drop("index",axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.describe()
```

### <font color= "blue"> Her satırdaki net_entries ve net_exits değerlerinin toplamı o turnikeden geçen toplam insan sayısını vereceği için total sütunu oluşturduk.

```python
df["total"] = df.net_entries + df.net_exits
```

```python
df
```

<font color="blue"> Her turnikeden geçen toplam insan saysını tutabilmek için "each_turnstile_data" verisetini oluşurup, her turnikeden farklı zamanlarda ve tarihlerde geçen insan sayısının toplamıyla yeni verisetimizi dolduruyoruz.

```python
each_turnstile_data = pd.DataFrame(df.groupby("turnstile").total.sum())
```

```python
each_turnstile_data.reset_index(inplace=True)
```

```python
each_turnstile_data
```

<font color="blue"> Turnikelerin istasyonlarını gösterecek sütunu ekleyelim.

```python
each_turnstile_data["station"] = [i.split("*")[-1] for i in each_turnstile_data.turnstile]
each_turnstile_data
```

<font color="blue"> Şimdi de bir duraktaki tüm turnikelerden geçen total kişi sayısını hesaplayıp yeni bir dataframe'de tutalım.

```python
each_station_data = pd.DataFrame(each_turnstile_data.groupby("station").total.sum())
each_station_data.reset_index(inplace=True)
each_station_data
```

### <font color="blue"> En yoğun  10 istasyonu çıkaralım.

```python
busiest_stations= pd.DataFrame(each_station_data.sort_values("total",ascending=False).head(10))
busiest_stations
```

```python
plt.figure(figsize=[15,7])
sns.barplot(x=busiest_stations.station,y=busiest_stations.total,palette="GnBu_r")
plt.xlabel("STATIONS",fontsize=16,weight="bold");
plt.xticks(ha="right",rotation=45,weight="bold")
plt.yticks(weight="bold")
plt.ylabel("TOTAL RIDERSHIP",fontsize=16,weight="bold")
plt.title("TOP TEN BUSIEST STATIONS IN NYC(2022)",fontsize=30,weight="bold");
```

### <font color="blue"> Saat dilimleri dengeli ve düzenli değil. Zaman aralıklarını düzenli hale getirmek için günü 4'er saat aralıklı "00:00,04:00,08:00,12:00,16:00,20:00"  olacak şekilde veri setimizi temizliyoruz.

```python
mask = (df_34_st_penn_sta.TIME != "00:00:00") & (df_34_st_penn_sta.TIME != "04:00:00") & (df_34_st_penn_sta.TIME != "08:00:00") & (df_34_st_penn_sta.TIME != "12:00:00") & (df_34_st_penn_sta.TIME != "16:00:00") & (df_34_st_penn_sta.TIME != "20:00:00")
df_34_st_penn_sta.drop(df_34_st_penn_sta[mask].index,axis=0,inplace=True)
df_34_st_penn_sta.reset_index(inplace=True)
```

```python
df_34_st_penn_sta.drop(["level_0","index"],axis=1,inplace=True)
```

```python
df_34_st_penn_sta
```

```python
df_34_st_penn_sta.groupby("day").TIME.value_counts()
```

<font color="blue"> Her saat dilimi için günlük veri sayısının birbirine yakın olmasından dolayı silme işlemimizin sakıncalı bir durum oluşturmadığını düşünerek bu veri seti üzerinde analizimizi genişletiyoruz.

```python
temp = pd.DataFrame(df_34_st_penn_sta.groupby(["day","TIME"]).total.mean())
temp.sort_values(by=["day","TIME"], inplace=True)
temp.reset_index(inplace=True)
```

```python
temp
```

```python
temp[temp.day=="Monday"].total.values
```

```python
day_and_time = pd.DataFrame({
    "TIME" : ["00:00","04:00","08:00","12:00","16:00","20:00"],
    "Monday": temp[temp.day=="Monday"].total.values,
    "Tuesday": temp[temp.day=="Tuesday"].total.values,
    "Wednesday": temp[temp.day=="Wednesday"].total.values,
    "Thursday": temp[temp.day=="Thursday"].total.values,
    "Friday": temp[temp.day=="Friday"].total.values,
    "Saturday": temp[temp.day=="Saturday"].total.values,
    "Sunday":temp[temp.day=="Sunday"].total.values
})
```

```python
day_and_time.set_index("TIME",inplace=True)
```

```python
day_and_time
```

```python
plt.figure(figsize=[10,7])
sns.heatmap(day_and_time,cmap="GnBu",linewidths=1,linecolor="w");
plt.xlabel("DAYS",fontsize=16,weight="bold");
plt.yticks(rotation= 0)
plt.ylabel("TIME",fontsize=16,weight="bold")
plt.title("BUSIEST DAY AND TIME AT 34-ST PENN STA",fontsize=20,weight="bold");
```

# <font color="blue"> Şimdi bir de gün ve saat analizini ayrı ayrı yapalım.

```python
temp
```

```python
#Günleri sıralayabilmek için yeni bir dataframe oluşturdum.
temp2 = pd.DataFrame(temp.groupby("day").total.mean())
days=[("Monday",1),("Tuesday",2),("Wednesday",3),("Thursday",4),("Friday",5),("Saturday",6),("Sunday",7)]
indexes=[]
for day in temp2.index:
    for i in range(7):
        if day == days[i][0]:
            indexes.append(days[i][1])
temp2["indexes"]=indexes
```

```python
temp2.sort_values(by="indexes",inplace=True)
temp2
```

# <font color="blue"> The most busy day is</font> <font color="dark pink"> Wednesday.

### <font color="blue"> Şimdi de çarşamba günü için saat analizi yapıyoruz.

```python
wednesday = pd.DataFrame(df_34_st_penn_sta[df_34_st_penn_sta.day=="Wednesday"])
```

```python
wed_time = wednesday.groupby("TIME").total.mean()
```

```python
# temp.groupby("day").total.sum()
plt.figure(figsize=[15,5])

plt.subplot(1,2,1)
sns.barplot(x=temp2.index,y=temp2.total,palette="GnBu_r")
plt.xlabel("Days",weight="bold")
plt.ylabel("Average Number of Ridership",weight="bold")
plt.title("AVERAGE NUMBER OF RIDERSHIP PER DAY",fontsize=12,weight="bold");

plt.subplot(1,2,2)
plt.pie(wed_time,colors=sns.color_palette("GnBu"),labels=wed_time.index,startangle=90,counterclock=False,explode=[0,0,0,0,.1,0]);
plt.title("AVERAGE RIDERSHIP OF WEDNESDAY BASED ON TIME",weight="bold",fontsize=12);
```
